## 선형회귀

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

## 시각화
# import matplotlib.pylab as pylab
import warnings
warnings.filterwarnings("ignore")
```

```python
def linear_regression(X_train, y_train, title):
    X_train = sm.add_constant(X_train)
    model = sm.OLS(y_train, X_train, axis = 1)
    model_trained = model.fit()
    
    model_residuals = model_trained.resid
    
    plt.rcParams['axes.unicode_minus'] = False
    fig,ax = plt.subplots(1, 1)
    fig.set_figheight(12)
    fig.set_figwidth(12)
    
    sm.ProbPlot(model_residuals).qqplot(line = 's', color = 'b', ax = ax)
#     sm.ProbPlot(model_residuals).qqplot(line='s', color='#1f77b4', ax=ax)
    ax.title.set_text(title)
    
    print(model_trained.summary())
```

## 데이터 분할

```python
from sklearn.model_selection import train_test_split
```

```python
def train_test(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X,
                                                       y,
                                                       train_size = 0.75,
                                                       random_state = 2023,
                                                       shuffle = True)
    
#     print(X_train.shape)
#     print(X_test.shape)
#     print(y_train.shape)
#     print(y_test.shape)
    
    linear_regression(X_train, y_train, y_train.columns)
```

```python
variables = apple.columns

for i in range(len(variables)):
    variables_new = variables.drop(variables[i])
    X1 = apple.loc[:, variables_new].values
    y1 = apple.loc[:, variables[i]].values
    
    X = pd.DataFrame(X1)
    y = pd.DataFrame(y1)
    
    X.columns = variables_new
    y.columns = [variables[i]]

    train_test(X, y)
```
